```python
path = '/content/gdrive/MyDrive/DataSet/train.csv'

import pandas as pd
BigMart = pd.read_csv(path)
```

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
BigMart.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(BigMart['Item_Weight'], BigMart['Item_Type'])
plt.xlabel('Item_Weight')
plt.ylabel('Item_Type')
plt.title('Item_Weight vs Item_Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
avg = BigMart.groupby('Item_Type')['Item_MRP'].mean().reset_index()

plt.plot( avg['Item_MRP'],avg['Item_Type'], marker = 'h',ms = '10' ,mec ='g', color = 'k', linestyle = '')
plt.ylabel('Item_MRP')
plt.xlabel('Item_Type')
plt.title('Avg(Item_MRP) vs Item_Type')
plt.show()
```
